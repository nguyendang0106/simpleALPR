2. Nội suy tuyến tính (Linear Interpolation)

- Lọc dữ liệu theo từng xe: Xác định car_id (ID của xe) để nhóm các khung hình liên quan đến từng xe. 

- Xác định các khoảng trống dữ liệu: Tìm các khoảng thời gian (khung hình) mà dữ liệu bounding box bị thiếu cho từng car_id. 

- Nội suy bounding box cho các khung hình bị thiếu: Trục thời gian (x): Khung hình hiện có tọa độ đầy đủ. 

- Dữ liệu cần nội suy (y): Tọa độ bounding box tại các khung hình hiện có. 

- Phương pháp: Dùng scipy.interpolate.interp1d với kind='linear' để nội suy các tọa độ (x1, y1, x2, y2) cho các khung hình bị thiếu. 

- Cách thực hiện nội suy: 

-Tạo một mảng các khung hình bị thiếu giữa 2 khung hình có dữ liệu (x_new). 

-Ước tính các bounding box bằng cách lấy trung bình tuyến tính giữa giá trị trước và sau (đường nối thẳng giữa hai giá trị). 

 

- Lợi ích của nội suy: 

 -Liên tục hóa dữ liệu: Duy trì thông tin đầy đủ cho tất cả các khung hình, ngay cả khi không có dữ liệu phát hiện từ mô hình. 

 -Cải thiện theo dõi: Nội suy giúp đảm bảo bounding box không bị ngắt quãng, giúp thuật toán SORT hoạt động mượt mà hơn. 

 -Dễ xử lý hậu kỳ: Dữ liệu đầy đủ hơn sẽ giúp các bước phân tích hoặc hiển thị video trực quan và mạch lạc hơn. 

 

Ví dụ trực quan: 

Giả sử bạn có các khung hình với tọa độ bounding box: 

Frame 1: [50, 50, 100, 100] 

Frame 4: [80, 70, 130, 120] 

Nếu khung hình 2 và 3 bị thiếu, nội suy sẽ tính toán: 

Frame 2: [60, 55, 110, 105] 

Frame 3: [70, 65, 120, 115] 

Đây là giá trị ước tính tuyến tính dựa trên các khung hình trước và sau.

### Chương trình
Chương trình của nhóm tôi đã thực hiện tốt các vấn đề như sau:
1. Xử lí biển số xe:
    
- Trong bài toán của bạn, cần phát hiện và đọc biển số xe: 

-Phát hiện biển số chỉ xảy ra trong một số khung hình nhất định khi biển số rõ ràng. 

-Theo dõi xe giúp bạn gán biển số đã nhận diện được với xe tương ứng, ngay cả khi biển số chỉ xuất hiện một lần. 

- Điều này giúp: 

-Đảm bảo mỗi biển số được gắn đúng với xe tương ứng. 

-Duy trì thông tin về xe ngay cả khi biển số không thể nhận diện trong một số khung hình.

2.Dự đoán vị trí tương lai 

- Theo dõi xe giúp dự đoán vị trí của xe trong các khung hình tiếp theo, ngay cả khi xe: 

-Bị che khuất một phần. 

-Tạm thời không được phát hiện bởi mô hình YOLO. 

- Điều này đặc biệt hữu ích trong các tình huống phức tạp, chẳng hạn xe bị khuất bởi cây cối hoặc các xe khác.

3.Tăng độ chính xác của hệ thống 

- Kết hợp giữa phát hiện vật thể (detection) và theo dõi vật thể (tracking) giúp hệ thống ổn định và chính xác hơn. 

-Detection chỉ hoạt động tốt trong từng khung hình riêng lẻ. 

-Tracking bổ sung mối liên kết giữa các khung hình, giúp hệ thống hoạt động liên tục và trơn tru.  

- Theo dõi xe trong video không chỉ giúp duy trì thông tin về từng xe qua các khung hình, mà còn cải thiện độ chính xác và khả năng ứng dụng của hệ thống trong thực tế. Đây là bước quan trọng khi làm việc với dữ liệu video liên tục.
